```python
import pandas as pd
import numpy as np
import os

class CheckeEmpty():
    
    def __init__(self, newfolder, oldfolder, year):
        self.newfolder = newfolder
        self.oldfolder = oldfolder
        self.year = year
        
    def empty(self):  #每次调取这个函数是一年的
        
        new_path = self.newfolder  #包含年份的一整条绝对路径
        old_path = self.oldfolder
        year = self. year
        # year = new_path[ new_path.rindex( '/' ) + 1 : len( new_path ) ] #截取年份
        
        new_datelist = os.listdir(new_path)   #仅遍历获取该日期下所有的时间文件夹名字的集合
        old_datelist = os.listdir(old_path)
        
        AllNewPath = []  #筛选过的完整路径集合
        AllOldPath = []      
        
        os.system("pwd") #确认是从yuqin work出发
        os.system("mkdir -p %s" %year)
        os.chdir("%s" %year)
        os.system("pwd")
        
        CheckEmpty_logname = "Check_Empty_" + str(year) + ".txt"
        CheckEmpty_log = open(CheckEmpty_logname , 'w')
        
        for n in range(len(AllNewPath)):
            
            getpath_new = AllNewPath[n]
            getpath_old = AllOldPath[n]

            print(getpath_new)
            print(getpath_old)
            
            
            # 开个log
            d = getpath_new
            time = d[ d.rindex( '/' ) + 1 : len( d ) ] 
            withoutLastStr = d[ 0 : d.rindex( '/' ) ]
            date= withoutLastStr[ withoutLastStr.rindex( '/' ) + 1 : len( withoutLastStr ) ]
            
            size_new = os.path.getsize(getpath_new)
            size_old = os.path.getsize(getpath_old)
            new_KB = float(size_new/1000)
            old_KB = float(size_old/1000)
            if new_KB < 400:
                CheckEmpty_log.write( date + time + ' \r\n')
            if old_KB < 400:
                CheckEmpty_log.write( date + time + ' \r\n')
            
        os.chdir("..")
            
        CheckEmpty_log.close
        print('CheckEmpty_log done')
            
            
def main():
    
    for year in range(2012, 2021):
#     year = 2015
        newfolder = "/dat/all/Equity/CN/marketdata/level2summary/5min/data/" + str(year)
        oldfolder= "/dat/all/Equity/CN/5min/data/" + str(year)
        print(newfolder)
        print(oldfolder)
        result = CheckeEmpty(newfolder, oldfolder, year)
        result.empty()
    
if __name__ == '__main__':
    
    main()

```
